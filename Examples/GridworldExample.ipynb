## Grid World Example

One archetype of MDPs suitable for Value Iteration are so called Gridworld Problems. In these, a player can move between positions on a rectangular grid. In this scenario, each position on the grid represents a state, and the cardinal directions in which they can move become the actions. We additionally assume that when choosing an action, there is a small probability that the player moves in a different available direction. Finally, by associating each position and movement with a reward based on where the player wishes to move, the problem becomes an MDP.

# A 2x2 Example.

Consider a 2x2 Grid with positions Top Left, Top Right, Bottom Left and Bottom Right (though these could equally be co-ordinates). The bottom right space is our target, and hence becomes an absorbing state. This means there are no available actions from this position. The Probabilities and Rewards associated with each move are given below.

When in the top left, we can move right or down.

| S  | a | s' | P(s'\|s,a) | R(s',s,a) |
|---|---|---|---|---|
| TL | R | TR | 0.9       | -1        |
| TL | R | BL | 0.1       | -2        |
| TL | D | BL | 0.9       | -2        |
| TL | D | TR | 0.1       | -1        |

When in the top right, we can move left or down.

| S  | a | s' | P(s'\|s,a) | R(s',s,a) |
|----|---|----|-----------|-----------|
| TR | L | TL | 0.9       | -3/2      |
| TR | L | BR | 0.1       | 10        |
| TR | D | BR | 0.8       | 15        |
| TR | D | TL | 0.2       | -1        |

When in the bottom left, we can move right or up.

| S  | a | s' | P(s'\|s,a) | R(s',s,a) |
|----|---|----|-----------|-----------|
| BL | R | BR | 0.9       | 20        |
| BL | R | TL | 0.1       | -5/2      |
| BL | U | TL | 0.8       | -1/2      |

And lastly, when in the bottom right position, there are no available moves. Note that the reward is related to entering the bottom right position.

To solve this with the Value Iteration package, we need to appropriately define each component.

```python
Grid_States = ["TL","TR","BL","BR"]

Grid_Actions = {"TL" : ["D","R"], "TR" : ["L","D"], "BL" : ["U","R"], "BR" : []}

Grid_Probs = {("TL","D") : [0,0.1,0.9,0], ("TL","R") : [0,0.9,0.1,0],
          ("TR","L") : [0.9,0,0,0.1], ("TR","D") : [0.2,0,0,0.8],
          ("BL","U") : [0.8,0,0,0.2], ("BL","R") : [0.1,0,0,0.9]}

Grid_Rewards = {("TL","D") : [0,-1,-2,0], ("TL","R") : [0,-1,-2,0],
          ("TR","L") : [-1.5,0,0,10], ("TR","D") : [-1,0,0,15],
          ("BL","U") : [-0.5,0,0,5], ("BL","R") : [-2.5,0,0,20]}

Grid_Discount = 0.8
```

We then choose a termination condition. Illustrated below are convergence of the value function and time-based termination.

```python
from ValueIteration import Convergence, TimeTerminate, Async_VI

Epsilon_Result = Async_VI(Grid_States, Grid_Actions, Grid_Probs, Grid_Rewards, Grid_Discount, Convergence(0.01))
print(Epsilon_Result)

Time_Result = Async_VI(Grid_States, Grid_Actions, Grid_Probs, Grid_Rewards, Grid_Discount, TimeTerminate(20))
print(Time_Result)
```

From this we see that for this simple gridworld problem, the optimal actions are to move down from top left, down from top right, and right from bottom left.
