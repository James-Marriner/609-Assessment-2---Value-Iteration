## Partying Example

In this simple example, we consider modelling an MDP which describes the health and actions of a partygoer. We suppose the partygoer can be healthy, or sick and must decide whether to party or relax. The state transition probabilities and rewards are:

| s | a | s' | P(s' \| s,a) | R(s,a) 
| ---|---|---|---|---|
| Healthy | Relax | Healthy | 0.95 | 7
| Healthy | Relax | Sick | 0.05 | 7
| Healthy | Party | Healthy | 0.7 |  10
| Healthy | Party | Sick | 0.3 | 10


| s | a | s' | P(s' \| s,a) | R(s,a) 
| ---|---|---|---|---|
| Sick | Relax | Healthy | 0.5 | 0
| Sick | Relax | Sick | 0.5 | 0
| Sick | Party | Healthy | 0.1 | 2 
| Sick | Party | Sick | 0.9 | 2

We specify these as the appropriate data structures, and choose a discount factor. Note that even though the expected rewards are specified above, we need to define these including s'.

```python
States = ["Healthy", "Sick"]

Actions = {"Healthy" : ["Relax","Party"],
           "Sick" : ["Relax", "Party"]}

Probs = {("Healthy","Relax") : [0.95,0.05],
         ("Healthy","Party") : [0.7,0.3],
         ("Sick","Relax") : [0.5,0.5],
         ("Sick", "Party") : [0.1,0.9]}

Rewards = {("Healthy","Relax") : [7,7],
           ("Healthy","Party") : [10,10],
           ("Sick","Relax") : [0,0],
           ("Sick", "Party") : [2,2]}

Discount = 0.9
```

The value iteration package offers multiple termination condiitons, two of which are specified below.

```python
from ValueIteration import Convergence, TimeTerminate, Async_VI

Epsilon_Result = Async_VI(States, Actions, Probs, Rewards, Discount, Convergence(0.01))
print(Epsilon_Result)

Time_Result = Async_VI(States, Actions, Probs, Rewards, Discount, TimeTerminate(5))
print(Time_Result)
```

From this we conclude it is best to party whilst helathy, and relax whilst sick.
